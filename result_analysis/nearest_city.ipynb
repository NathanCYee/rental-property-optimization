```python
# Import all libraries

import pandas as pd
import numpy as np
import glob
import os
import pickle

from sklearn.neighbors import KNeighborsClassifier
```

```python
model = pickle.load(open('../model_creation/outputs/city_classifier/model.pkl', 'rb'))
```

```python
def match(path):
    # load csv and predict 
    frame = pd.read_csv(path,index_col=0)
    params = frame[['income','density']]
    result = pd.DataFrame(model.predict(params), columns=['city'])
    #concatenate and return
    newframe = pd.concat([frame, result], axis=1)
    return newframe
```

```python
def process(path):
    result = match(path)
    display(result)
    # display unique cities in the dataset
    display(result['city'].unique())
    # output back to csv
    ind = path.rindex('.') # get file separator
    result.to_csv(path[:ind]+'_transformed.csv')
    
```

```python
paths = ['../optimization/outputs/rent_sale_tbe/moead.csv',
         '../optimization/outputs/rent_sale_tbe/smpso.csv',
         '../optimization/outputs/rent_sale_tbe/nsga2.csv',
         '../optimization/outputs/rent_sale/moead.csv',
         '../optimization/outputs/rent_sale/smpso.csv',
         '../optimization/outputs/rent_sale/nsga2.csv']
```

```python
for path in paths:
    print(path)
    process(path)
```
