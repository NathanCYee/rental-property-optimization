# Describe data and run dbscan on income/density
Want to do some unsupervised exploration of the pareto front.  
Using code from [scikit-learn examples](https://scikit-learn.org/stable/auto_examples/cluster/plot_dbscan.html)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
framelist = ['../optimization/outputs/rent_sale_tbe/moead.csv',
             '../optimization/outputs/rent_sale_tbe/smpso.csv',
             '../optimization/outputs/rent_sale_tbe/nsga2.csv',
             '../optimization/outputs/rent_sale/moead.csv',
             '../optimization/outputs/rent_sale/smpso.csv',
             '../optimization/outputs/rent_sale/nsga2.csv']
```

```python
for path in framelist:
    print(path)
    frame = pd.read_csv(path,index_col=0)
    display(frame.describe())
    scaler = StandardScaler()
    x = frame[['income', 'density']]
    X = scaler.fit_transform(x)
    db = DBSCAN(eps=0.5, min_samples=2).fit(X)
    core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
    core_samples_mask[db.core_sample_indices_] = True
    labels = db.labels_

    # Number of clusters in labels, ignoring noise if present.
    n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise_ = list(labels).count(-1)

    print("Estimated number of clusters: %d" % n_clusters_)
    print("Estimated number of noise points: %d" % n_noise_)
    print("Silhouette Coefficient: %0.3f" % metrics.silhouette_score(X, labels))
    # Black removed and is used for noise instead.
    unique_labels = set(labels)
    fig, ax = plt.subplots()
    colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

    for k, col in zip(unique_labels, colors):
        if k == -1:
            # Black used for noise.
            col = [0, 0, 0, 1]

        class_member_mask = labels == k
        # plot points in the cluster and a core sample
        xy = X[class_member_mask & core_samples_mask]
        if len(xy) != 0:
            xy = scaler.inverse_transform(xy)
        plt.plot(
            xy[:, 0],
            xy[:, 1],
            "o",
            markerfacecolor=tuple(col),
            markeredgecolor="k",
            markersize=14,
        )

        # plot non-samples
        xy = X[class_member_mask & ~core_samples_mask]
        if len(xy) != 0:
            xy = scaler.inverse_transform(xy)
        plt.plot(
            xy[:, 0],
            xy[:, 1],
            "o",
            markerfacecolor=tuple(col),
            markeredgecolor="k",
            markersize=6,
        )

    ax.set_xlabel("income")
    ax.set_ylabel("density")
    plt.title("Estimated number of clusters: %d" % n_clusters_)
    plt.show()
```
