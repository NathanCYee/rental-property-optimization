```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os
```

```python
def get_correlation(path):
    """displays a correlation matrix for a rental/sale dataset"""
    # load data
    parquet_files = glob.glob(os.path.join(path, "*.parquet"))
    data = pd.concat((pd.read_parquet(f) for f in parquet_files))
    # generate graph
    fig, ax = plt.subplots()
    c = ax.matshow(data.corr(), interpolation='nearest')
    fig.colorbar(c)
    ax.set_xticklabels(['']+list(data.columns))
    ax.set_yticklabels(['']+list(data.columns))
    plt.show()
```

```python
get_correlation('../data_engineering/outputs/rental/')
```

```python
get_correlation('../data_engineering/outputs/sales/')
```
